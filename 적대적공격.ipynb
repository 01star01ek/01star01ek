```python
#@title 단일 프로세스 프레임 추출 코드
import cv2
import os
from glob import glob
from tqdm import tqdm

video_dir = "/content/input_videos"  #@param {type:"string"}
output_base = "/content/extract_frames"  #@param {type:"string"}
extract_per_sec = 7  #@param {type:"integer"}

os.makedirs(output_base, exist_ok=True)

def extract_video_frames(video_path):
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    output_dir = os.path.join(output_base, video_name)
    os.makedirs(output_dir, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    interval = max(1, int(fps // extract_per_sec))

    frame_idx = 0
    frame_list = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if frame_idx % interval == 0:
            frame_list.append(frame)
        frame_idx += 1

    cap.release()

    # 여기서 한 번에 저장!
    for saved_idx, frame in enumerate(frame_list):
        frame_path = os.path.join(output_dir, f"key_{saved_idx:04d}.jpg")
        cv2.imwrite(frame_path, frame)

    return f"{video_name}: {len(frame_list)} frames saved"

# 실행
video_paths = glob(os.path.join(video_dir, "*.mp4"))
print(f"🎬 총 {len(video_paths)}개의 영상 처리 시작")

for video_path in tqdm(video_paths, desc="📦 Processing videos"):
    msg = extract_video_frames(video_path)
```

```python
#@title 피치 제한 강화 + 눈뜸 50% 기준 + KeyError 해결 코드
import cv2, os, math, numpy as np, pandas as pd
from glob import glob
from tqdm import tqdm
import mediapipe as mp

# MediaPipe 초기화
mp_face_mesh = mp.solutions.face_mesh
FACE_MESH = mp_face_mesh.FaceMesh(
    static_image_mode=True,
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.7,
    min_tracking_confidence=0.5
)

# 눈 좌표 인덱스
LEFT_EYE_IDX = [
    33, 7, 163, 144, 145, 153, 154, 155, 133, 173, 157, 158, 159, 160, 161, 246,
    33, 173, 157, 158, 159, 160, 161, 246, 33
]
RIGHT_EYE_IDX = [
    362, 398, 384, 385, 386, 387, 388, 466, 263, 249, 390, 373, 374, 380, 381, 382,
    362, 398, 384, 385, 386, 387, 388, 466, 362
]

CORE_LEFT_EYE = [33, 160, 158, 133, 153, 144]
CORE_RIGHT_EYE = [362, 385, 387, 263, 373, 380]
LEFT_EYE_VERTICAL = [159, 145]
RIGHT_EYE_VERTICAL = [386, 374]
POSE_IDX = [1, 152, 33, 263, 61, 291]

# 파라미터 설정 - 피치 제한 강화, 눈뜸 기준 완화
YAW_T = 25
PITCH_T = 25  # 35에서 25로 강화 (피치 관대하게 하지 않음)
HEAD_DOWN_BONUS = 1  # 보너스 최소화
ENABLE_ADAPTIVE_EAR = True
EAR_PERCENTILE_HIGH = 37 # 상위 50%를 눈뜬 상태로 판정 (기존 40%에서 완화)
EAR_PERCENTILE_LOW = 10

model_points = np.array([
    (0.0, 0.0, 0.0),
    (0.0, -330.0, -65.0),
    (-225.0, 170.0, -135.0),
    (225.0, 170.0, -135.0),
    (-150.0, -150.0, -125.0),
    (150.0, -150.0, -125.0)
], dtype="double")

def get_mediapipe_landmarks(img):
    h, w = img.shape[:2]
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    res = FACE_MESH.process(rgb)
    if not res.multi_face_landmarks:
        return None
    lm = res.multi_face_landmarks[0]
    coords = np.array([[p.x * w, p.y * h] for p in lm.landmark])
    return coords

def estimate_pose_mediapipe(landmarks, img_shape):
    image_points = landmarks[POSE_IDX]
    focal = img_shape[1]
    center = (img_shape[1]/2, img_shape[0]/2)
    cam = np.array([[focal, 0, center[0]], [0, focal, center[1]], [0, 0, 1]], dtype="double")
    dist = np.zeros((4,1))
    success, rv, _ = cv2.solvePnP(model_points, image_points, cam, dist)
    return rv if success else None

def rotation_vector_to_euler(rv):
    rmat, _ = cv2.Rodrigues(rv)
    proj = np.hstack((rmat, np.zeros((3,1))))
    angles = cv2.decomposeProjectionMatrix(proj)[6]
    pitch = math.degrees(math.asin(math.sin(math.radians(angles[1][0]))))
    yaw   = math.degrees(math.asin(math.sin(math.radians(angles[2][0]))))
    roll  = -math.degrees(math.asin(math.sin(math.radians(angles[0][0]))))
    return pitch, yaw, roll

def eye_aspect_ratio(eye):
    A = np.linalg.norm(eye[1] - eye[5])
    B = np.linalg.norm(eye[2] - eye[4])
    C = np.linalg.norm(eye[0] - eye[3])
    return (A + B) / (2.0 * C)

def enhanced_eye_aspect_ratio(landmarks, is_left=True):
    if is_left:
        outer = landmarks[33]
        inner = landmarks[133]
        v1 = np.linalg.norm(landmarks[159] - landmarks[145])
        v2 = np.linalg.norm(landmarks[158] - landmarks[153])
        v3 = np.linalg.norm(landmarks[160] - landmarks[144])
    else:
        outer = landmarks[362]
        inner = landmarks[263]
        v1 = np.linalg.norm(landmarks[386] - landmarks[374])
        v2 = np.linalg.norm(landmarks[385] - landmarks[373])
        v3 = np.linalg.norm(landmarks[387] - landmarks[380])

    horizontal = np.linalg.norm(outer - inner)
    avg_vertical = (v1 + v2 + v3) / 3.0
    return avg_vertical / horizontal

def calculate_adaptive_ear_thresholds(all_ear_values):
    """영상별 적응형 EAR 임계값 계산 - 50% 기준 적용"""
    if len(all_ear_values) < 5:
        return {
            'high_threshold': 0.23,
            'medium_threshold': 0.18,
            'low_threshold': 0.15,
            'min_threshold': 0.12
        }

    ear_array = np.array(all_ear_values)

    # 50% 기준으로 임계값 설정
    high_threshold = np.percentile(ear_array, 100 - EAR_PERCENTILE_HIGH)  # 상위 50%
    medium_threshold = np.percentile(ear_array, 100 - EAR_PERCENTILE_HIGH * 1.2)  # 상위 60%
    low_threshold = np.percentile(ear_array, 100 - EAR_PERCENTILE_HIGH * 1.5)  # 상위 75%
    min_threshold = np.percentile(ear_array, EAR_PERCENTILE_LOW)  # 하위 10%

    # 최소값 보장
    high_threshold = max(high_threshold, 0.16)  # 더 관대하게
    medium_threshold = max(medium_threshold, 0.13)
    low_threshold = max(low_threshold, 0.10)
    min_threshold = max(min_threshold, 0.07)

    return {
        'high_threshold': high_threshold,
        'medium_threshold': medium_threshold,
        'low_threshold': low_threshold,
        'min_threshold': min_threshold,
        'ear_stats': {
            'mean': np.mean(ear_array),
            'std': np.std(ear_array),
            'min': np.min(ear_array),
            'max': np.max(ear_array),
            'count': len(ear_array)
        }
    }

def is_eye_open_adaptive(landmarks, thresholds, pitch_angle=0):
    """적응형 EAR 기반 눈뜸 판정 - 피치 조정 최소화"""

    left_eye_basic = landmarks[CORE_LEFT_EYE]
    right_eye_basic = landmarks[CORE_RIGHT_EYE]

    basic_left_ear = eye_aspect_ratio(left_eye_basic)
    basic_right_ear = eye_aspect_ratio(right_eye_basic)
    basic_avg_ear = (basic_left_ear + basic_right_ear) / 2.0

    enhanced_left_ear = enhanced_eye_aspect_ratio(landmarks, True)
    enhanced_right_ear = enhanced_eye_aspect_ratio(landmarks, False)
    enhanced_avg_ear = (enhanced_left_ear + enhanced_right_ear) / 2.0

    ear_difference = abs(basic_left_ear - basic_right_ear)

    # 피치 조정 최소화 (관대하게 하지 않음)
    pitch_factor = 1.0
    if pitch_angle > 20:  # 20도 이상에서만 최소 조정
        pitch_factor = max(0.95, 1.0 - (pitch_angle - 20) * 0.005)  # 최대 5%만 완화

    # 적응형 임계값 적용
    adj_high = thresholds['high_threshold'] * pitch_factor
    adj_medium = thresholds['medium_threshold'] * pitch_factor
    adj_low = thresholds['low_threshold'] * pitch_factor
    adj_min = thresholds['min_threshold'] * pitch_factor

    # 4단계 눈뜸 판정
    level_1 = (basic_avg_ear > adj_high and
               enhanced_avg_ear > adj_high * 0.9 and
               basic_left_ear > adj_medium and
               basic_right_ear > adj_medium and
               ear_difference < 0.08)

    level_2 = (basic_avg_ear > adj_medium and
               enhanced_avg_ear > adj_medium * 0.8 and
               basic_left_ear > adj_low and
               basic_right_ear > adj_low and
               ear_difference < 0.12)

    level_3 = (basic_avg_ear > adj_low and
               basic_left_ear > adj_min and
               basic_right_ear > adj_min and
               ear_difference < 0.15)

    level_4 = (basic_avg_ear > adj_min and
               basic_left_ear > adj_min * 0.8 and
               basic_right_ear > adj_min * 0.8)

    # 레벨별 점수 부여
    if level_1:
        eye_level = 4
    elif level_2:
        eye_level = 3
    elif level_3:
        eye_level = 2
    elif level_4:
        eye_level = 1
    else:
        eye_level = 0

    return eye_level, {
        'basic_ear': basic_avg_ear,
        'enhanced_ear': enhanced_avg_ear,
        'left_ear': basic_left_ear,
        'right_ear': basic_right_ear,
        'ear_diff': ear_difference,
        'eye_level': eye_level,
        'pitch_factor': pitch_factor,
        'thresholds_used': {
            'high': adj_high,
            'medium': adj_medium,
            'low': adj_low,
            'min': adj_min
        }
    }

def frontal_score_strict_pitch(c):
    """피치 제한 강화된 정면성 평가 함수"""

    yaw_angle = abs(c['yaw'])
    pitch_angle = c['pitch']

    # 엄격한 각도 제한 (피치 관대하게 하지 않음)
    if yaw_angle > YAW_T:  # 25도
        return -1000
    if abs(pitch_angle) > PITCH_T:  # ±25도 (엄격)
        return -1000

    # 기본 페널티 (피치에 더 큰 가중치)
    yaw_penalty = yaw_angle * 0.8
    pitch_penalty = abs(pitch_angle) * 1.2  # 피치 페널티 증가

    # 고개 숙임 보너스 최소화
    head_down_bonus = 0
    if -15 <= pitch_angle <= -5:  # 아주 제한적인 범위에서만
        head_down_bonus = HEAD_DOWN_BONUS * 0.5  # 보너스도 절반으로

    # 눈뜸 레벨 보너스
    eye_level = c.get('eye_level', 0)
    eye_bonus = eye_level * 12  # 눈뜸이 더 중요

    bonus = eye_bonus + (head_down_bonus if pitch_angle < 0 else 0)

    return -(0.5 * yaw_penalty + 0.5 * pitch_penalty) + bonus

def calculate_final_quality_score(pitch, yaw, eye_level, ear_details):
    """최종 품질 점수 계산 - 피치 페널티 강화"""

    # 각도 점수 (피치에 더 큰 페널티)
    yaw_score = max(0, 100 - abs(yaw) * 2.0)
    pitch_score = max(0, 100 - abs(pitch) * 2.5)  # 피치 페널티 증가
    angle_score = (yaw_score + pitch_score) / 2

    # 눈뜸 레벨 점수 (50% 기준이므로 더 관대)
    eye_score = eye_level * 25

    # EAR 품질 점수
    ear_quality = min(100, ear_details['basic_ear'] * 300)

    # 종합 점수 (눈뜸 비중 증가)
    total_score = (
        angle_score * 0.3 +    # 각도 30%
        eye_score * 0.5 +      # 눈뜸 50% (증가)
        ear_quality * 0.2      # EAR 품질 20%
    )

    return total_score

# 경로 설정
INPUT_ROOT = "/content/extract_frames"
OUTPUT_ROOT = "/content/alignmented_frame"
os.makedirs(OUTPUT_ROOT, exist_ok=True)

file_ext = ".jpg"
video_dirs = [d for d in os.listdir(INPUT_ROOT) if os.path.isdir(os.path.join(INPUT_ROOT, d))]
missing_videos = []
detailed_log = []
best_images = []
best_names = []

print(f"📦 총 {len(video_dirs)}개 영상 처리 시작")
print(f"🎯 눈뜸 기준: 상위 {EAR_PERCENTILE_HIGH}% (50% 기준으로 완화)")
print(f"📐 각도 제한: Yaw ±{YAW_T}°, Pitch ±{PITCH_T}° (피치 제한 강화)")
print(f"🎁 고개 숙임 보너스: {HEAD_DOWN_BONUS}점 (최소화)")

for video_name in tqdm(video_dirs, desc="🎯 Strict pitch + 50% eye threshold"):
    input_dir = os.path.join(INPUT_ROOT, video_name)

    # 1단계: 모든 프레임의 EAR 값 수집
    all_ear_values = []
    frame_data = []

    for f in sorted(os.listdir(input_dir)):
        if not f.lower().endswith(file_ext):
            continue

        full_path = os.path.join(input_dir, f)
        img = cv2.imread(full_path)
        if img is None:
            continue

        lm = get_mediapipe_landmarks(img)
        if lm is None:
            continue

        rv = estimate_pose_mediapipe(lm, img.shape)
        if rv is None:
            continue

        pitch, yaw, _ = rotation_vector_to_euler(rv)

        # 엄격한 각도 제한
        if abs(yaw) > YAW_T * 1.5 or abs(pitch) > PITCH_T * 1.5:
            continue

        # EAR 계산
        left_eye = lm[CORE_LEFT_EYE]
        right_eye = lm[CORE_RIGHT_EYE]
        left_ear = eye_aspect_ratio(left_eye)
        right_ear = eye_aspect_ratio(right_eye)
        avg_ear = (left_ear + right_ear) / 2.0

        all_ear_values.append(avg_ear)

        # 프레임 데이터 저장
        x1, y1 = lm[:,0].min(), lm[:,1].min()
        x2, y2 = lm[:,0].max(), lm[:,1].max()
        face_area = (x2 - x1) * (y2 - y1)
        cx, cy = (x1 + x2) / 2, (y1 + y2) / 2

        frame_data.append({
            'img': img,
            'filename': f,
            'landmarks': lm,
            'pitch': pitch,
            'yaw': yaw,
            'avg_ear': avg_ear,
            'cx': cx,
            'cy': cy,
            'face_area': face_area
        })

    if not all_ear_values:
        missing_videos.append(video_name)
        detailed_log.append({
            'video_name': video_name,
            'total_frames': 0,
            'selected': False,
            'selection_level': 0,  # KeyError 방지를 위해 추가
            'reason': 'No valid frames found'
        })
        continue

    # 2단계: 적응형 임계값 계산 (50% 기준)
    thresholds = calculate_adaptive_ear_thresholds(all_ear_values)

    # 3단계: 4단계 후보 분류
    level_4_candidates = []
    level_3_candidates = []
    level_2_candidates = []
    level_1_candidates = []

    for frame in frame_data:
        # 엄격한 피치 제한 적용
        if abs(frame['pitch']) > PITCH_T:
            continue

        eye_level, eye_details = is_eye_open_adaptive(
            frame['landmarks'], thresholds, abs(frame['pitch'])
        )

        if eye_level == 0:
            continue

        candidate = {
            'img': frame['img'],
            'filename': frame['filename'],
            'pitch': frame['pitch'],
            'yaw': frame['yaw'],
            'cx': frame['cx'],
            'cy': frame['cy'],
            'face_area': frame['face_area'],
            'w': frame['img'].shape[1],
            'h': frame['img'].shape[0],
            'eye_level': eye_level,
            'eye_details': eye_details,
            'quality_score': calculate_final_quality_score(
                frame['pitch'], frame['yaw'], eye_level, eye_details
            )
        }

        if eye_level == 4:
            level_4_candidates.append(candidate)
        elif eye_level == 3:
            level_3_candidates.append(candidate)
        elif eye_level == 2:
            level_2_candidates.append(candidate)
        else:
            level_1_candidates.append(candidate)

    # 4단계: 최적 프레임 선택
    best_img = None
    best_filename = None
    selection_reason = "No suitable frames"
    selection_level = 0

    if level_4_candidates:
        best = max(level_4_candidates, key=frontal_score_strict_pitch)
        best_img = best['img']
        best_filename = best['filename']
        selection_reason = f"Level 4: Selected from {len(level_4_candidates)} highest quality candidates"
        selection_level = 4

    elif level_3_candidates:
        best = max(level_3_candidates, key=frontal_score_strict_pitch)
        best_img = best['img']
        best_filename = best['filename']
        selection_reason = f"Level 3: Selected from {len(level_3_candidates)} high quality candidates"
        selection_level = 3

    elif level_2_candidates:
        best = max(level_2_candidates, key=frontal_score_strict_pitch)
        best_img = best['img']
        best_filename = best['filename']
        selection_reason = f"Level 2: Selected from {len(level_2_candidates)} medium quality candidates"
        selection_level = 2

    elif level_1_candidates:
        best = max(level_1_candidates, key=lambda x: x['quality_score'])
        best_img = best['img']
        best_filename = best['filename']
        selection_reason = f"Level 1: Selected from {len(level_1_candidates)} minimum quality candidates"
        selection_level = 1

    elif frame_data:
        # 최후의 수단: 가장 품질 좋은 프레임 무조건 선택
        best_frame = max(frame_data, key=lambda x: x['avg_ear'])
        best_img = best_frame['img']
        best_filename = best_frame['filename']
        selection_reason = f"Emergency: Selected best EAR frame ({best_frame['avg_ear']:.3f})"
        selection_level = 0

    # KeyError 방지를 위한 안전한 로그 기록
    detailed_log.append({
        'video_name': video_name,
        'total_frames': len(frame_data),
        'level_4_candidates': len(level_4_candidates),
        'level_3_candidates': len(level_3_candidates),
        'level_2_candidates': len(level_2_candidates),
        'level_1_candidates': len(level_1_candidates),
        'selected': best_filename is not None,
        'selection_level': selection_level,  # 항상 포함
        'best_filename': best_filename,
        'reason': selection_reason,
        'ear_thresholds': thresholds,
        'quality_score': best.get('quality_score', 0) if 'best' in locals() else 0
    })

    # 프레임 저장
    if best_img is not None:
        save_path = os.path.join(OUTPUT_ROOT, f"{video_name}.jpg")
        cv2.imwrite(save_path, best_img)

        if selection_level <= 1:
            print(f"⚠️  {video_name}: Low quality selection (Level {selection_level})")
    else:
        missing_videos.append(video_name)

# 결과 저장
if missing_videos:
    df_missing = pd.DataFrame(missing_videos, columns=["video_name"])
    df_missing.to_csv("no_frame_found.csv", index=False)
    print(f"❗ {len(missing_videos)}개 영상에서 프레임을 찾지 못함")

df_log = pd.DataFrame(detailed_log)
df_log.to_csv("strict_pitch_50percent_eye_log.csv", index=False)
print(f"📊 처리 결과 저장: strict_pitch_50percent_eye_log.csv")

# KeyError 방지된 안전한 통계 출력
success_rate = (len(video_dirs) - len(missing_videos)) / len(video_dirs) * 100 if video_dirs else 0

# .get() 메서드 사용으로 KeyError 방지
level_4_usage = sum(1 for log in detailed_log if log.get('selection_level', 0) == 4)
level_3_usage = sum(1 for log in detailed_log if log.get('selection_level', 0) == 3)
level_2_usage = sum(1 for log in detailed_log if log.get('selection_level', 0) == 2)
level_1_usage = sum(1 for log in detailed_log if log.get('selection_level', 0) == 1)
emergency_usage = sum(1 for log in detailed_log if log.get('selection_level', 0) == 0)

print(f"✅ 성공률: {success_rate:.1f}% ({len(video_dirs) - len(missing_videos)}/{len(video_dirs)})")
print(f"🏆 Level 4 (최고품질): {level_4_usage}개")
print(f"🥈 Level 3 (고품질): {level_3_usage}개")
print(f"🥉 Level 2 (중품질): {level_2_usage}개")
print(f"📉 Level 1 (최소품질): {level_1_usage}개")
print(f"🚨 Emergency (강제선택): {emergency_usage}개")

print(f"\n📈 품질 분포:")
print(f"   - 고품질 이상 (Level 3+): {(level_4_usage + level_3_usage) / len(video_dirs) * 100:.1f}%")
print(f"   - 중품질 이상 (Level 2+): {(level_4_usage + level_3_usage + level_2_usage) / len(video_dirs) * 100:.1f}%")
print(f"   - 최소품질 이상 (Level 1+): {(len(video_dirs) - emergency_usage) / len(video_dirs) * 100:.1f}%")

print(f"\n🎯 설정 요약:")
print(f"   - 피치 제한: ±{PITCH_T}° (엄격)")
print(f"   - 눈뜸 기준: 상위 {EAR_PERCENTILE_HIGH}% (관대)")
print(f"   - 고개 숙임 보너스: {HEAD_DOWN_BONUS}점 (최소)")
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# 1. Load latent dictionary
latent_path = "/content/experiments/restyle_e4e_sg3/latents.npy"
latent_dict = np.load(latent_path, allow_pickle=True).item()

# 2. 정렬된 파일 리스트 확보
filenames = sorted(latent_dict.keys())

# 3. 각 latent에서 마지막 step → 평균 → (512,)
latents = []
for key in filenames:
    latent = latent_dict[key][-1]  # 마지막 step (18, 512)
    mean_latent = latent.mean(axis=0).astype('float32')  # (512,)
    latents.append(mean_latent)

latents = np.stack(latents)  # shape: (N, 512)

# 4. cosine similarity matrix
sim_matrix = cosine_similarity(latents)  # shape: (N, N)

# 5. 각 query 파일마다 top-3 유사한 match + score 저장
rows = []

for i in range(len(filenames)):
    sims = sim_matrix[i].copy()
    sims[i] = -np.inf  # 자기 자신 제외
    top3_idx = np.argsort(sims)[::-1][:3]
    row = {
        "query": filenames[i],
        "top1": filenames[top3_idx[0]],
        "top2": filenames[top3_idx[1]],
        "top3": filenames[top3_idx[2]],
        "top1val": round(float(sims[top3_idx[0]]), 6),
        "top2val": round(float(sims[top3_idx[1]]), 6),
        "top3val": round(float(sims[top3_idx[2]]), 6),
    }
    rows.append(row)

# 6. Save to CSV
df = pd.DataFrame(rows)
df.to_csv("video_similarity_top3_compact.csv", index=False)

print("✅ Saved to video_similarity_top3_compact.csv")
```

```python
# ========== SimSwap 완전 새로 시작 파이프라인 ==========
# neuralchen/SimSwap 기반

import cv2
import subprocess
import os
from glob import glob
import sys
import warnings
import re
from pathlib import Path
from tqdm import tqdm
import shutil

# 로그 정리
warnings.filterwarnings('ignore')
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ['PYTHONWARNINGS'] = 'ignore'

class FreshSimSwapPipeline:
    def __init__(self):
        """처음부터 새로 만든 SimSwap 파이프라인"""
        print("🚀 SimSwap 파이프라인 새로 시작!")
        self.install_and_setup()

    def install_and_setup(self):
        """SimSwap 완전 새로 설치 및 설정"""
        print("🔧 SimSwap 완전 새로 설치 중...")

        # 1. 기존 SimSwap 완전 제거
        if os.path.exists('/content/SimSwap'):
            print("   기존 SimSwap 제거 중...")
            shutil.rmtree('/content/SimSwap')

        # 2. 필수 패키지 설치
        print("   📦 필수 패키지 설치 중...")
        os.system('pip install -q torch==2.0.1 torchvision==0.15.2 --index-url https://download.pytorch.org/whl/cu118')
        os.system('pip install -q insightface==0.7.3 onnxruntime moviepy opencv-python imageio==2.34.0')

        # 3. 올바른 SimSwap 클론
        print("   📥 neuralchen/SimSwap 클론 중...")
        os.chdir('/content')
        os.system('git clone -q https://github.com/neuralchen/SimSwap.git')
        os.chdir('/content/SimSwap')

        # 4. 모델 다운로드
        self.download_models()

        # 5. 호환성 수정
        self.fix_compatibility()

        # 6. 경로 설정 (상대 경로 사용)
        self.arcface_path = './arcface_model/arcface_checkpoint.tar'  # 상대 경로로 변경
        self.checkpoints_dir = './checkpoints'  # 상대 경로로 변경

        print("✅ SimSwap 설치 및 설정 완료!")

    def download_models(self):
        """모든 필요한 모델 다운로드"""
        print("   📥 모델 다운로드 중...")

        # 디렉토리 생성
        dirs_to_create = [
            '/content/arcface_model',
            '/content/SimSwap/parsing_model/checkpoint',
            '/content/SimSwap/checkpoints',
            '/content/SimSwap/insightface_func/models'
        ]

        for dir_path in dirs_to_create:
            os.makedirs(dir_path, exist_ok=True)

        # 모델 다운로드
        downloads = [
            ('arcface 모델', 'https://github.com/woctezuma/SimSwap-colab/releases/download/1.0/arcface_checkpoint.tar', '/content/arcface_model/arcface_checkpoint.tar'),
            ('checkpoints', 'https://github.com/neuralchen/SimSwap/releases/download/1.0/checkpoints.zip', '/content/checkpoints.zip'),
            ('parsing 모델', 'https://github.com/neuralchen/SimSwap/releases/download/1.0/79999_iter.pth', '/content/SimSwap/parsing_model/checkpoint/79999_iter.pth'),
            ('antelope 모델', 'https://github.com/woctezuma/SimSwap-colab/releases/download/antelope/antelope.zip', '/content/antelope.zip')
        ]

        for name, url, output in downloads:
            print(f"     다운로드 중: {name}")
            os.system(f'wget -q {url} -O {output}')

        # 압축 해제
        print("     압축 해제 중...")
        os.system('unzip -q /content/checkpoints.zip -d /content/SimSwap/checkpoints')
        os.system('unzip -q /content/antelope.zip -d /content/SimSwap/insightface_func/models/')

        # arcface 압축 해제 (SimSwap 디렉토리 내에)
        os.chdir('/content/SimSwap')
        os.makedirs('arcface_model', exist_ok=True)
        os.system('cp /content/arcface_model/arcface_checkpoint.tar ./arcface_model/')
        os.chdir('arcface_model')
        os.system('tar -xf arcface_checkpoint.tar')
        os.chdir('/content/SimSwap')

        # 추가: checkpoints 구조 확인 및 수정
        print("     모델 구조 확인 중...")
        if os.path.exists('/content/SimSwap/checkpoints'):
            checkpoint_files = os.listdir('/content/SimSwap/checkpoints')
            print(f"     checkpoints 내용: {checkpoint_files}")

            # people 폴더가 없으면 생성하고 파일 이동
            people_dir = '/content/SimSwap/checkpoints/people'
            if not os.path.exists(people_dir):
                os.makedirs(people_dir, exist_ok=True)
                # .pth 파일들을 people 폴더로 이동
                for file in checkpoint_files:
                    if file.endswith('.pth'):
                        src = f'/content/SimSwap/checkpoints/{file}'
                        dst = f'/content/SimSwap/checkpoints/people/{file}'
                        os.system(f'cp {src} {dst}')
                        print(f"     {file} -> people/")

    def fix_compatibility(self):
        """호환성 문제 수정"""
        print("   🔧 호환성 수정 중...")

        # 1. PyTorch 호환성 수정
        if os.path.exists('models/fs_model.py'):
            os.system("sed -i.bak 's/torch.load(netArc_checkpoint, map_location=torch.device(\"cpu\"))/torch.load(netArc_checkpoint, map_location=torch.device(\"cpu\"), weights_only=False)/' models/fs_model.py")

        # 2. 감지 크기 수정 (더 작게 해서 얼굴 감지율 향상)
        if os.path.exists('test_wholeimage_swapsingle.py'):
            os.system("sed -i 's/det_size=(640,640)/det_size=(320,320)/' test_wholeimage_swapsingle.py")
            # crop_size도 조정
            os.system("sed -i 's/crop_size=224/crop_size=512/' test_wholeimage_swapsingle.py")

        # 3. NMS 임계값 수정 (더 관대하게)
        if os.path.exists('test_wholeimage_swapsingle.py'):
            with open('test_wholeimage_swapsingle.py', 'r') as f:
                content = f.read()

            if 'app.models["detection"].nms_thresh' not in content:
                new_lines = []
                for line in content.splitlines():
                    new_lines.append(line)
                    if 'app.prepare(' in line:
                        new_lines.append('    if hasattr(app.models, "detection"):')
                        new_lines.append('        app.models["detection"].nms_thresh = 0.5')  # 더 관대하게
                        new_lines.append('        app.models["detection"].det_thresh = 0.4')   # 더 관대하게

                with open('test_wholeimage_swapsingle.py', 'w') as f:
                    f.write('\n'.join(new_lines))

        # 4. arcface 경로 수정 (상대 경로로)
        if os.path.exists('test_wholeimage_swapsingle.py'):
            with open('test_wholeimage_swapsingle.py', 'r') as f:
                content = f.read()

            # Arc_path 기본값을 상대 경로로 수정
            content = content.replace(
                "parser.add_argument('--Arc_path', type=str, default='arcface_model/arcface_checkpoint.tar'",
                "parser.add_argument('--Arc_path', type=str, default='./arcface_model/arcface_checkpoint.tar'"
            )

            with open('test_wholeimage_swapsingle.py', 'w') as f:
                f.write(content)

    def check_face_in_image(self, image_path):
        """이미지에서 얼굴 감지 확인"""
        try:
            import cv2
            img = cv2.imread(image_path)
            if img is None:
                return False, "이미지를 읽을 수 없음"

            # 이미지 크기 확인
            height, width = img.shape[:2]
            if width < 100 or height < 100:
                return False, f"이미지가 너무 작음: {width}x{height}"

            # 얼굴이 있는지 간단 체크 (OpenCV 얼굴 감지)
            face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, 1.3, 5)

            if len(faces) > 0:
                return True, f"얼굴 감지됨: {len(faces)}개"
            else:
                return False, "얼굴이 감지되지 않음"

        except Exception as e:
            return False, f"검사 중 오류: {e}"

    def preprocess_image_for_face_detection(self, image_path, output_path):
        """얼굴 감지를 위한 이미지 전처리"""
        try:
            import cv2
            img = cv2.imread(image_path)
            if img is None:
                return False

            # 이미지 크기 조정 (너무 크거나 작으면 감지 어려움)
            height, width = img.shape[:2]

            # 적절한 크기로 조정 (512-1024 사이)
            if width > 1024 or height > 1024:
                scale = 1024 / max(width, height)
                new_width = int(width * scale)
                new_height = int(height * scale)
                img = cv2.resize(img, (new_width, new_height), interpolation=cv2.INTER_AREA)
            elif width < 512 and height < 512:
                scale = 512 / max(width, height)
                new_width = int(width * scale)
                new_height = int(height * scale)
                img = cv2.resize(img, (new_width, new_height), interpolation=cv2.INTER_CUBIC)

            # 밝기/대비 조정
            lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
            l, a, b = cv2.split(lab)
            clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
            l = clahe.apply(l)
            enhanced = cv2.merge([l, a, b])
            img = cv2.cvtColor(enhanced, cv2.COLOR_LAB2BGR)

            cv2.imwrite(output_path, img)
            return True

        except Exception as e:
            print(f"     ⚠️ 이미지 전처리 실패: {e}")
            return False

    def extract_number_from_filename(self, filename):
        """파일명에서 숫자 추출"""
        numbers = re.findall(r'\d+', os.path.splitext(filename)[0])
        return int(numbers[0]) if numbers else None

    def find_video_image_pairs(self, video_dir, image_dir):
        """영상과 이미지 쌍 찾기"""
        print(f"🔍 매칭 쌍 찾는 중...")
        print(f"   영상 디렉토리: {video_dir}")
        print(f"   이미지 디렉토리: {image_dir}")

        # 파일 찾기
        video_files = glob(os.path.join(video_dir, "*.mp4"))
        image_files = glob(os.path.join(image_dir, "*.jpg")) + glob(os.path.join(image_dir, "*.png"))

        print(f"   발견된 영상: {len(video_files)}개")
        print(f"   발견된 이미지: {len(image_files)}개")

        # 번호별 매칭
        video_dict = {}
        for video_path in video_files:
            number = self.extract_number_from_filename(os.path.basename(video_path))
            if number is not None:
                video_dict[number] = video_path

        image_dict = {}
        for image_path in image_files:
            number = self.extract_number_from_filename(os.path.basename(image_path))
            if number is not None:
                image_dict[number] = image_path

        # 매칭되는 쌍 생성
        pairs = []
        for number in sorted(set(video_dict.keys()) & set(image_dict.keys())):
            pairs.append({
                'number': number,
                'video_path': video_dict[number],
                'image_path': image_dict[number],
                'video_name': os.path.splitext(os.path.basename(video_dict[number]))[0],
                'image_name': os.path.splitext(os.path.basename(image_dict[number]))[0]
            })

        return pairs

    def extract_frames_from_video(self, video_path, frame_dir, interval=5):
        """비디오에서 프레임 추출"""
        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"❌ 비디오를 열 수 없음: {video_path}")
            return None, None

        # 비디오 정보
        original_fps = cap.get(cv2.CAP_PROP_FPS)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        new_fps = original_fps / interval

        os.makedirs(frame_dir, exist_ok=True)

        frame_count = 0
        saved_count = 0

        print(f"     원본 FPS: {original_fps:.1f}, 총 프레임: {total_frames}")
        print(f"     간격: {interval}, 새 FPS: {new_fps:.1f}")

        with tqdm(total=total_frames, desc="     프레임 추출") as pbar:
            while True:
                ret, frame = cap.read()
                if not ret:
                    break

                if frame_count % interval == 0:
                    frame_path = os.path.join(frame_dir, f'frame_{saved_count:04d}.jpg')
                    cv2.imwrite(frame_path, frame)
                    saved_count += 1

                frame_count += 1
                pbar.update(1)

        cap.release()
        print(f"     ✅ {saved_count}개 프레임 추출 완료")
        return original_fps, new_fps

    def run_simswap_processing(self, source_img, frame_dir, swap_dir):
        """SimSwap으로 얼굴 교체 처리"""
        frames = sorted([f for f in os.listdir(frame_dir) if f.endswith('.jpg')])
        total_frames = len(frames)

        if total_frames == 0:
            print("❌ 처리할 프레임이 없음")
            return False

        os.makedirs(swap_dir, exist_ok=True)

        # 소스 이미지 얼굴 검사
        face_ok, face_msg = self.check_face_in_image(source_img)
        print(f"     소스 이미지 얼굴 검사: {face_msg}")

        # 얼굴이 감지되지 않으면 전처리 시도
        processed_source = source_img
        if not face_ok:
            print("     소스 이미지 전처리 시도 중...")
            processed_source = source_img.replace('.jpg', '_processed.jpg').replace('.png', '_processed.png')
            if self.preprocess_image_for_face_detection(source_img, processed_source):
                face_ok, face_msg = self.check_face_in_image(processed_source)
                print(f"     전처리 후 얼굴 검사: {face_msg}")

        success_count = 0

        print(f"     총 {total_frames}개 프레임 처리 시작")

        # SimSwap 디렉토리로 이동하여 실행
        original_cwd = os.getcwd()
        os.chdir('/content/SimSwap')

        with tqdm(total=total_frames, desc="     SimSwap 처리") as pbar:
            for i, frame_name in enumerate(frames):
                target_img = os.path.join(frame_dir, frame_name)
                result_subdir = os.path.join(swap_dir, f'frame_{i:04d}')
                os.makedirs(result_subdir, exist_ok=True)

                # 절대 경로로 변환 (SimSwap 디렉토리에서 실행하므로)
                abs_source = os.path.abspath(processed_source)
                abs_target = os.path.abspath(target_img)
                abs_output = os.path.abspath(result_subdir)

                # SimSwap 명령어 (절대 경로 사용)
                cmd = [
                      'python', 'test_wholeimage_swapsingle.py',
                      '--crop_size', '512',
                      '--use_mask',
                      '--no_simswaplogo',
                      '--name', 'people',
                      '--Arc_path', './arcface_model/arcface_checkpoint.tar',
                      '--pic_a_path', abs_source,
                      '--pic_b_path', abs_target,
                      '--pic_specific_path', abs_target,
                      '--output_path', abs_output,
                      '--checkpoints_dir', './checkpoints',
                      '--which_epoch', 'latest'   # ⭐ 추가됨
                  ]

                # 첫 번째 프레임만 에러 체크
                if i == 0:
                    print(f"\n     첫 프레임 테스트 중...")
                    print(f"     작업 디렉토리: {os.getcwd()}")
                    print(f"     소스: {abs_source}")
                    print(f"     타겟: {abs_target}")

                    # 필수 파일들 존재 확인
                    essential_files = [
                        './arcface_model/arcface_checkpoint.tar',
                        './checkpoints/people',
                        './test_wholeimage_swapsingle.py'
                    ]

                    for file_path in essential_files:
                        if os.path.exists(file_path):
                            print(f"     ✅ {file_path} 존재")
                        else:
                            print(f"     ❌ {file_path} 없음")

                    # people 폴더 내용 확인
                    if os.path.exists('./checkpoints/people'):
                        people_files = os.listdir('./checkpoints/people')
                        print(f"     people 폴더: {people_files[:3]}...")

                    result = subprocess.run(cmd, capture_output=True, text=True)
                    if result.returncode != 0:
                        print(f"\n❌ SimSwap 첫 프레임 실패:")
                        print(f"   명령어: {' '.join(cmd)}")
                        print(f"   에러: {result.stderr[:400]}...")
                        print(f"   출력: {result.stdout[:400]}...")

                        # 추가 디버깅 정보
                        print(f"   현재 디렉토리 내용:")
                        current_files = os.listdir('.')
                        for f in current_files[:10]:
                            print(f"     {f}")

                        os.chdir(original_cwd)
                        return False

                    # 결과 확인
                    result_files = [f for f in os.listdir(abs_output) if f.endswith(('.jpg', '.png'))]
                    if not result_files:
                        print(f"\n❌ 결과 파일이 생성되지 않음")
                        os.chdir(original_cwd)
                        return False

                    success_count += 1
                    print(f"     ✅ 첫 프레임 성공: {result_files[0]}")
                else:
                    # 나머지는 조용히 처리
                    subprocess.run(cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

                    # 성공 확인
                    result_files = [f for f in os.listdir(abs_output) if f.endswith(('.jpg', '.png'))]
                    if result_files:
                        success_count += 1

                pbar.update(1)

        # 원래 디렉토리로 복귀
        os.chdir(original_cwd)

        print(f"     ✅ 성공: {success_count}/{total_frames} 프레임")
        return success_count > 0

    def create_video_from_results(self, swap_dir, output_video, fps):
        """결과 이미지들로 비디오 생성"""
        print(f"     비디오 생성 중: {fps:.1f}fps")

        # 결과 이미지 수집
        result_images = []
        swap_subdirs = sorted([d for d in os.listdir(swap_dir)
                              if os.path.isdir(os.path.join(swap_dir, d))])

        for subdir in swap_subdirs:
            subdir_path = os.path.join(swap_dir, subdir)
            result_files = [f for f in os.listdir(subdir_path)
                           if f.endswith(('.jpg', '.png'))]
            if result_files:
                result_images.append(os.path.join(subdir_path, result_files[0]))

        if not result_images:
            print("❌ 결과 이미지가 없음")
            return False

        # 첫 번째 이미지로 비디오 설정 확인
        first_img = cv2.imread(result_images[0])
        if first_img is None:
            print("❌ 첫 번째 이미지를 읽을 수 없음")
            return False

        height, width, _ = first_img.shape
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        out = cv2.VideoWriter(output_video, fourcc, fps, (width, height))

        if not out.isOpened():
            print("❌ 비디오 라이터 열기 실패")
            return False

        # 비디오 생성
        with tqdm(total=len(result_images), desc="     비디오 생성") as pbar:
            for img_path in result_images:
                img = cv2.imread(img_path)
                if img is not None:
                    out.write(img)
                pbar.update(1)

        out.release()
        print(f"     ✅ 비디오 생성 완료: {output_video}")
        return True

    def interpolate_video_fps(self, input_video, output_video, target_fps):
        """FFmpeg로 프레임 보간"""
        print(f"     프레임 보간: → {target_fps:.1f}fps")

        cmd = [
            'ffmpeg', '-i', input_video,
            '-filter:v', f'minterpolate=fps={target_fps}:mi_mode=mci:mc_mode=aobmc:vsbmc=1',
            '-c:v', 'libx264', '-crf', '23', '-preset', 'medium',
            '-y', output_video
        ]

        try:
            result = subprocess.run(cmd, capture_output=True, text=True)
            if result.returncode == 0:
                print(f"     ✅ 보간 완료: {output_video}")
                return True
            else:
                print(f"     ❌ 보간 실패")
                return False
        except Exception as e:
            print(f"     ❌ 보간 오류: {e}")
            return False

    def process_single_pair(self, video_path, image_path, output_dir, frame_interval=5):
        """단일 영상-이미지 쌍 처리"""
        print(f"  📹 영상: {os.path.basename(video_path)}")
        print(f"  🖼️ 이미지: {os.path.basename(image_path)}")

        # 디렉토리 설정
        frame_dir = os.path.join(output_dir, "frames")
        swap_dir = os.path.join(output_dir, "swapped")

        # 1. 프레임 추출
        print("  1️⃣ 프레임 추출 중...")
        original_fps, new_fps = self.extract_frames_from_video(video_path, frame_dir, frame_interval)
        if original_fps is None:
            return False

        # 2. SimSwap 처리
        print("  2️⃣ SimSwap 얼굴 교체 중...")
        if not self.run_simswap_processing(image_path, frame_dir, swap_dir):
            return False

        # 3. 기본 비디오 생성
        print("  3️⃣ 기본 비디오 생성 중...")
        base_video = os.path.join(output_dir, "base_video.mp4")
        if not self.create_video_from_results(swap_dir, base_video, new_fps):
            return False

        return True, original_fps, base_video

    def process_all_pairs(self, video_dir="/content/input_videos",
                         image_dir="/content/experiments/restyle_e4e_sg3/inference_results/0",
                         output_dir="/content/simswap_results",
                         frame_interval=5,
                         use_interpolation=True):
        """모든 쌍 처리"""

        # 매칭 쌍 찾기
        pairs = self.find_video_image_pairs(video_dir, image_dir)

        if not pairs:
            print("❌ 매칭되는 쌍을 찾을 수 없습니다!")
            return

        print(f"🎯 총 {len(pairs)}개 쌍 발견!")
        os.makedirs(output_dir, exist_ok=True)

        success_count = 0

        for i, pair in enumerate(pairs, 1):
            print(f"\n🚀 [{i}/{len(pairs)}] 처리 중...")

            # 출력 디렉토리
            pair_name = f"pair_{pair['number']:03d}_{pair['video_name']}"
            pair_output_dir = os.path.join(output_dir, pair_name)

            try:
                # 단일 쌍 처리
                result = self.process_single_pair(
                    pair['video_path'],
                    pair['image_path'],
                    pair_output_dir,
                    frame_interval
                )

                if not result:
                    print(f"  ❌ 쌍 {pair['number']} 실패")
                    continue

                success, original_fps, base_video = result

                # 4. 프레임 보간 (선택사항)
                if use_interpolation:
                    print("  4️⃣ 프레임 보간 중...")
                    final_video = os.path.join(pair_output_dir, f"{pair_name}_final.mp4")

                    if self.interpolate_video_fps(base_video, final_video, original_fps):
                        print(f"  🎉 최종 완료: {final_video}")
                    else:
                        print(f"  ⚠️ 보간 실패, 기본 사용: {base_video}")
                else:
                    print(f"  🎉 완료: {base_video}")

                success_count += 1

            except Exception as e:
                print(f"  ❌ 오류: {e}")
                continue

        print(f"\n🎉 전체 완료: {success_count}/{len(pairs)}개 성공!")
        print(f"📁 결과 위치: {output_dir}")

# ========== 사용 방법 ==========
if __name__ == "__main__":
    # 파이프라인 초기화 (자동으로 설치 및 설정)
    pipeline = FreshSimSwapPipeline()

    # 처리 시작
    pipeline.process_all_pairs(
        video_dir="/content/input_videos",
        image_dir="/content/experiments/restyle_e4e_sg3/inference_results/0",
        output_dir="/content/simswap_results",
        frame_interval=5,
        use_interpolation=True
    )
```

```python
# AdversarialDeepFakes 직접 영상 공격 파이프라인
# 완성된 SimSwap 영상에 바로 적대적 노이즈 추가

import os
import subprocess
import sys
from glob import glob
from pathlib import Path
import warnings

warnings.filterwarnings('ignore')

class DirectVideoAttacker:
    def __init__(self):
        """AdversarialDeepFakes 직접 영상 공격기"""
        print("🎯 AdversarialDeepFakes 직접 영상 공격기 초기화")
        self.setup_adversarial_deepfakes()

    def setup_adversarial_deepfakes(self):
        """AdversarialDeepFakes 환경 설정"""
        print("📦 AdversarialDeepFakes 설정 중...")

        # 1. 레포지토리 클론
        if not os.path.exists('/content/AdversarialDeepFakes'):
            print("   📥 AdversarialDeepFakes 클론 중...")
            os.chdir('/content')
            os.system('git clone -q https://github.com/paarthneekhara/AdversarialDeepFakes.git')
        else:
            print("   ✅ AdversarialDeepFakes 이미 존재")

        os.chdir('/content/AdversarialDeepFakes')

        # 2. 필요한 패키지 설치
        print("   📦 필요 패키지 설치 중...")

        # requirements.txt가 있으면 설치
        if os.path.exists('requirements.txt'):
            os.system('pip install -q -r requirements.txt')

        # 기본 필수 패키지들
        essential_packages = [
            'torch', 'torchvision', 'opencv-python', 'numpy',
            'pillow', 'matplotlib', 'scikit-learn', 'tqdm', 'argparse'
        ]

        for package in essential_packages:
            os.system(f'pip install -q {package}')

        # 3. 디렉토리 및 파일 구조 확인
        print("   📁 레포지토리 구조 확인:")
        files = os.listdir('.')
        important_files = ['attack.py', 'detect_from_video.py', 'models.py']

        for file in important_files:
            if file in files:
                print(f"     ✅ {file} 존재")
            else:
                print(f"     ❌ {file} 없음")

        # 4. 사전 훈련된 모델 다운로드 (필요한 경우)
        self.download_pretrained_models()

        print("✅ AdversarialDeepFakes 설정 완료")

    def download_pretrained_models(self):
        """사전 훈련된 딥페이크 탐지 모델 다운로드"""
        print("   📥 사전 훈련된 모델 다운로드 중...")

        # 모델 디렉토리 생성
        os.makedirs('models', exist_ok=True)
        os.makedirs('weights', exist_ok=True)

        # 일반적인 딥페이크 탐지 모델들 (예시)
        model_urls = [
            # Xception 기반 모델 (FaceForensics++ 데이터셋 훈련)
            ('xception_model.pth', 'https://github.com/ondyari/FaceForensics/releases/download/v1.0/xception-b5690688.pth'),
            # MesoNet 모델
            ('mesonet_model.pth', 'https://github.com/DariusAf/MesoNet/releases/download/v1.0/mesonet.pth')
        ]

        # 실제 사용 가능한 모델이 있는지 확인하고 다운로드
        for model_name, url in model_urls:
            model_path = f'weights/{model_name}'
            if not os.path.exists(model_path):
                print(f"     다운로드 시도: {model_name}")
                result = subprocess.run(['wget', '-q', url, '-O', model_path], capture_output=True)
                if result.returncode == 0:
                    print(f"     ✅ {model_name} 다운로드 완료")
                else:
                    print(f"     ⚠️ {model_name} 다운로드 실패 (더미 모델 사용)")
                    # 더미 모델 생성
                    self.create_dummy_model(model_path)
            else:
                print(f"     ✅ {model_name} 이미 존재")

    def create_dummy_model(self, model_path):
        """더미 모델 생성 (실제 모델이 없는 경우)"""
        import torch
        import torch.nn as nn

        class DummyDetector(nn.Module):
            def __init__(self):
                super().__init__()
                self.features = nn.Sequential(
                    nn.Conv2d(3, 32, 3, padding=1),
                    nn.ReLU(),
                    nn.AdaptiveAvgPool2d(1),
                    nn.Flatten(),
                    nn.Linear(32, 2)
                )

            def forward(self, x):
                return self.features(x)

        dummy_model = DummyDetector()
        torch.save(dummy_model.state_dict(), model_path)
        print(f"     🔧 더미 모델 생성: {model_path}")

    def find_completed_videos(self, simswap_results_dir="/content/simswap_results"):
        """완성된 SimSwap 영상들 찾기"""
        print(f"🔍 완성된 SimSwap 영상 찾는 중: {simswap_results_dir}")

        video_files = []

        # _final.mp4 파일들 먼저 찾기 (보간된 최종 결과)
        final_videos = glob(os.path.join(simswap_results_dir, "**/*_final.mp4"), recursive=True)

        # base_video.mp4 파일들 찾기 (기본 결과)
        base_videos = glob(os.path.join(simswap_results_dir, "**/base_video.mp4"), recursive=True)

        # 모든 mp4 파일 찾기
        all_videos = glob(os.path.join(simswap_results_dir, "**/*.mp4"), recursive=True)

        # 우선순위: final > base > 기타
        for video in final_videos:
            video_files.append(('final', video))

        for video in base_videos:
            if video not in [v[1] for v in video_files]:
                video_files.append(('base', video))

        for video in all_videos:
            if video not in [v[1] for v in video_files]:
                video_files.append(('other', video))

        print(f"   발견된 영상: {len(video_files)}개")
        for vtype, vpath in video_files:
            print(f"     {vtype}: {os.path.basename(vpath)}")

        return video_files

    def run_adversarial_attack(self, input_video, output_video, attack_type='robust', model_type='xception'):
        """AdversarialDeepFakes attack.py 실행"""
        print(f"\n🎯 적대적 공격 실행")
        print(f"   입력: {os.path.basename(input_video)}")
        print(f"   출력: {os.path.basename(output_video)}")
        print(f"   공격 타입: {attack_type}")
        print(f"   모델 타입: {model_type}")

        # 모델 경로 설정
        model_weights = f'weights/{model_type}_model.pth'
        if not os.path.exists(model_weights):
            model_weights = f'weights/xception_model.pth'  # 기본값

        # 출력 디렉토리 생성
        output_dir = os.path.dirname(output_video)
        os.makedirs(output_dir, exist_ok=True)

        # attack.py 명령어 구성
        cmd = [
            'python', 'attack.py',
            '-i', input_video,
            '-mi', model_weights,
            '-mt', model_type,
            '-o', output_dir,
            '-a', attack_type
        ]

        # CUDA 사용 가능하면 추가
        import torch
        if torch.cuda.is_available():
            cmd.append('--cuda')

        print(f"   명령어: {' '.join(cmd)}")

        try:
            # attack.py 실행
            result = subprocess.run(cmd, capture_output=True, text=True, cwd='/content/AdversarialDeepFakes')

            if result.returncode == 0:
                print(f"   ✅ 공격 성공!")
                print(f"   출력: {result.stdout[:200]}...")

                # 생성된 파일 확인
                output_files = glob(os.path.join(output_dir, "*.mp4")) + glob(os.path.join(output_dir, "*.avi"))
                if output_files:
                    # 생성된 파일을 원하는 이름으로 이동
                    generated_file = output_files[0]
                    os.rename(generated_file, output_video)
                    print(f"   📁 결과 파일: {output_video}")
                    return True
                else:
                    print(f"   ❌ 출력 파일이 생성되지 않음")
                    return False
            else:
                print(f"   ❌ 공격 실패:")
                print(f"   stderr: {result.stderr[:300]}")
                print(f"   stdout: {result.stdout[:300]}")
                return False

        except Exception as e:
            print(f"   ❌ 실행 중 오류: {e}")
            return False

    def process_all_videos(self, simswap_results_dir="/content/simswap_results",
                          output_dir="/content/adversarial_results",
                          attack_types=['robust', 'iterative_fgsm'],
                          model_type='xception'):
        """모든 완성된 영상에 적대적 공격 적용"""

        # 완성된 영상들 찾기
        video_files = self.find_completed_videos(simswap_results_dir)

        if not video_files:
            print("❌ 처리할 영상이 없습니다!")
            return

        print(f"\n🚀 총 {len(video_files)}개 영상 × {len(attack_types)}개 공격 = {len(video_files) * len(attack_types)}개 작업")

        os.makedirs(output_dir, exist_ok=True)
        success_count = 0
        total_tasks = len(video_files) * len(attack_types)

        for i, (vtype, video_path) in enumerate(video_files, 1):
            video_name = os.path.splitext(os.path.basename(video_path))[0]

            print(f"\n[{i}/{len(video_files)}] 처리 중: {video_name}")

            for j, attack_type in enumerate(attack_types, 1):
                print(f"\n  🎯 공격 [{j}/{len(attack_types)}]: {attack_type}")

                # 출력 파일명 생성
                output_filename = f"{video_name}_{attack_type}_adversarial.mp4"
                output_path = os.path.join(output_dir, output_filename)

                # 이미 존재하면 스킵
                if os.path.exists(output_path):
                    print(f"     ⚠️ 이미 존재함: {output_filename}")
                    success_count += 1
                    continue

                # 적대적 공격 실행
                if self.run_adversarial_attack(video_path, output_path, attack_type, model_type):
                    success_count += 1
                    print(f"     🎉 완료: {output_filename}")
                else:
                    print(f"     ❌ 실패: {output_filename}")

        print(f"\n🎉 전체 완료: {success_count}/{total_tasks}개 성공!")
        print(f"📁 결과 위치: {output_dir}")

        # 결과 요약
        if success_count > 0:
            print(f"\n📊 생성된 적대적 영상들:")
            result_files = glob(os.path.join(output_dir, "*.mp4"))
            for rf in result_files:
                print(f"   🎬 {os.path.basename(rf)}")

# ========== 사용 방법 ==========
if __name__ == "__main__":
    # 적대적 공격기 초기화
    attacker = DirectVideoAttacker()

    # 모든 완성된 SimSwap 영상에 노이즈 추가
    attacker.process_all_videos(
        simswap_results_dir="/content/simswap_results",
        output_dir="/content/adversarial_results",
        attack_types=['robust', 'iterative_fgsm', 'black_box'],  # 원하는 공격 타입들
        model_type='xception'  # 'xception' 또는 'meso'
    )

    print("\n✨ AdversarialDeepFakes 노이즈 추가 완료!")
    print("   완성된 SimSwap 영상들이 다양한 적대적 노이즈와 함께 생성되었습니다.")
```
